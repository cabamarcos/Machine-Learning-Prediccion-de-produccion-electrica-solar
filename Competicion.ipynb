```python
# Importamos las librer√≠as correspondientes
import pandas as pd
import numpy as np
import pickle


# Importamos nuestro modelo
with open('modelo_final.pkl', "rb") as file:
    modelo_final = pickle.load(file)

comp_df = pd.read_csv("comp_st27ns1.txt.bz2", compression="bz2", index_col=0)

cols_to_drop=['dlwrf_s2_1', 'dlwrf_s4_1', 'tcdc_ea1_1', 'tcdc_ea3_1', 'tcdc_ea5_1', 'tcolc_e2_1',
              'tcolc_e4_1', 'tmp_2m_4_1', 'tmax_2m3_1', 'tmax_2m4_1', 'tmin_2m1_1', 'tmp_sfc1_1',
              'ulwrf_s5_1', 'ulwrf_s2_1', 'ulwrf_s3_1', 'ulwrf_s1_1', 'tmax_2m2_1', 'tmax_2m5_1',
              'tmin_2m2_1', 'tmin_2m3_1', 'tmin_2m5_1', 'ulwrf_s4_1', 'tmp_sfc5_1', 'tmp_sfc2_1']

# Elimina las columnas especificadas
comp_df_reduced = comp_df.drop(cols_to_drop, axis=1)



# Pedimos al usuario que introduzca un area y calcular
# su precio usando el modelo importado
pred = modelo_final.predict(comp_df_reduced)

print(f"predicciones: {pred}")
np.savetxt("predicciones.csv", pred, delimiter=' ')
```
